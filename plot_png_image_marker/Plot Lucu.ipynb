```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import image
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

path = "315A39F4-705A-48C3-A7CD-592EBF49AB50-removebg-preview.png"
image = plt.imread(path)#[116:116+30, 236:236+30]

plt.style.use ('ggplot')

x = list(map(str, np.arange(20)))
y = np.random.randint(low = 20, high = 100, size = 20)

fig, ax = plt.subplots()
ax.plot(x,y, 'b--')

def plot_images(x, y, image, ax=None):
    ax = ax or plt.gca()

    for xi, yi in zip(x,y):
        im = OffsetImage(image, zoom=8/ax.figure.dpi)
        im.image.axes = ax

        ab = AnnotationBbox(im, (xi,yi), frameon=False, pad=0.0,)

        ax.add_artist(ab)

plot_images(x, y, image, ax=ax)

fig = plt.gcf()
fig.set_size_inches(12, 10)

ax.set_title("Rage!!!!\n", fontsize = 20, color = 'red')
ax.set_ylabel("Rage Level!!!! ", fontsize = 15, color = 'blue')
ax.set_xlabel("Day", fontsize = 15, color = 'b')
plt.show()
```
